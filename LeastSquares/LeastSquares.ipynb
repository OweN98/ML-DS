# Least Squares最小二乘法  
* 由(xi, yi)(i = 1, 2, 3, ...)拟合出h(x)，误差为h(xi) - yi，当L2范数最小时，拟合最好。  
* H(x) = w0 + w1 * x + w2 * x^2 + w3 * x^3 + ... + wn * x^n，wn(n = 0, 1, 2, ...)是参数。  
* 最小二乘法的目的就是找到一组wn，使得L2范数最小。  
* The aim of Least Squares is to find the best group of *wn* which makes L2 the least. And we get the best fitting.

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy.optimize import leastsq
```

```python
def fx(x):
    return np.sin(2 * np.pi * x)

def fitting_fx(w, x):
    f = np.poly1d(w)
    return f(x)

def loss(w, x, y):
    loss_w = fitting_fx(w, x) - y
    return loss_w
```

```python
x = np.linspace(0, 1, 10)
x_points = np.linspace(0, 1, 1000)

y_points = fx(x)
y = [np.random.normal(0, 0.1) + y1 for y1 in y_points]

def go_fitting(M):
    w_init = np.random.rand(M + 1)
    #least squares
    w_lq = leastsq(loss, w_init, args=(x, y))
    print('Fitting Parameters: ', w_lq[0])
    
    plt.plot(x_points, fx(x_points), label = 'fx')
    plt.plot(x_points, fitting_fx(w_lq[0], x_points), label = 'fitting')
    plt.plot(x, y, 'bo', label = 'noise')
    plt.legend()
    
    return w_lq

def regularization(w, x, y):
    reg = fitting_fx(w, x) - y
    reg = np.append(reg, np.sqrt(0.5 * 0.0001 * np.square(w)))
    return reg

def regularied_fitting(M):
    w_init = np.random.rand(M + 1)
    w_lq = leastsq(loss, w_init, args = (x, y))
    w_lq_reg = leastsq(regularization, w_init, args=(x, y))
    plt.plot(x_points, fx(x_points), label = 'fx')
    plt.plot(x_points, fitting_fx(w_lq[0], x_points), label = 'fitting')
    plt.plot(x_points, fitting_fx(w_lq_reg[0], x_points), label = 'regularization' )
    plt.plot(x, y, 'bo', label = 'noise')
    plt.legend()
    
    return w_lq
```

```python
w_lq_1 = go_fitting(0)
```
